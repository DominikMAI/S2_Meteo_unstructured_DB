```python
###  Konfiguracja klienta do zapytań z witryny Creodias.
setup_logging(verbose = 1)
dag = EODataAccessGateway(user_conf_file_path = 'eodag_sentinel.yml')
```

```python
# Limit maksylnej dopuszczalnego przesłonięcia terenu chmurami
cloud_coverage_limit = 80

#Zakres dni objętych odfiltrowaniem
day_delta_minus = 2
day_delta_plus = 3

# Zasięg obszaru zainteresowania - Polska
extent = {"lonmin":14, "latmin": 48, "lonmax": 24, "latmax": 55}

products, nr_of_search = dag.search(productType='S2_MSI_L1C', 
                                    items_per_page = 200,
                                    geom = extent,
                                    start = "2020-01-01",
                                    end = "2021-03-31",
                                    cloudCover = cloud_coverage_limit)

g = products.as_geojson_object()
```

### 1.1. Przekształcenie wyników zapytania w ramkę danych.

```python
f = pd.json_normalize(g['features'])
S2_df = pd.DataFrame([])
```

```python
# Identyfikator zobrazowania
S2_df["id"] = f['id']

# Geometria poligonu zasięgu zdjecia
S2_df["geometry"] = f['geometry.coordinates']
S2_df['geometry'] = S2_df['geometry'].apply(lambda x: Polygon(x[0])).astype(str)

# Rodzaj instrumetu pozyskującego zdjęcia S2_MSI_L1C
S2_df["product_type"] = f['properties.eodag_product_type']

# Link do pobrania zdjęcia z zasobu Creodias
S2_df["url"] = f['properties.services.download.url']

# Poziom prztworzenia np. czy została wykonana korekcja atmosferyczna
S2_df["processingLevel"] = f['properties.processingLevel']

# Procentowe pokrycie chmurami, przesłaniającymi teren
S2_df["cloudCover"] = f['properties.cloudCover']

# Procentowe pokrycie śniegiem, przesłaniającym teren
S2_df["snow_cover"] = f['properties.snowCover']
S2_df["publicationDate"] = f['properties.publicationDate']

# Współrzędne środka masy poligonu zasięgu zdjęcia
S2_df["centroid"] = f['properties.centroid.coordinates']
S2_df['publicationDate'] = S2_df['publicationDate'].apply(lambda x: x.split('T')[0].replace("-", ""))
```

### 2. Pobranie danych pogodowych sieci obserwacji METAR
#### Data pobierana jest z wyników 

```python
def Get_Weather_Measurements_Data(date):
    """
    Funkcja przeprowadza web-scraping serwisu uniwersytetu Wyyoming(US) z danymi pogodowymi sieci METAR (http://weather.uwyo.edu).
    """
    station = ['EPBY','EPGD','EPGD','EPKK','EPKT','EPLB', 'EPMO','EPPO','EPRZ','EPSC','EPSY','EPWA','EPWR']
    weather_stations = []
    godzina = 14
    
    for stn in station:
        r = requests.get(f"http://weather.uwyo.edu/cgi-bin/wyowx.fcgi?TYPE=current&DATE={date}&HOUR={godzina}Z&UNITS=M&STATION={stn}")
        soup = BeautifulSoup(r.text,'lxml')

        location = soup.find('h5').get_text().split(" ")
        city_name = " ".join(soup.find('h3').get_text().split("Observations for")).split(",")[0]

        lat = float(location[3].split("E")[0])
        lon = float(location[1].split("N")[0])

        df = pd.read_html(soup.prettify())[0].T
        df.columns = df.apply(lambda x: x.astype(str).values[0]+ '[' + x.astype(str) .values[2] + ']')
        df = df.iloc[1 : 2, :7]

        df['lat'] = lat
        df['lon'] = lon
        df['city_name'] = city_name
        df['station'] = stn
        df['date'] = date
        weather_stations.append(df)
        
    weather_stations_df = pd.concat(weather_stations) 
    return weather_stations_df
```

### 3. Etap uzupełnienia danych o zasięgu obrazów satelitarnych atrybutami warunków pogodowych z nabliższej stacji pogodowej.

```python
def distance(pt_1, pt_2):
    """
    Funkcja wyznaczająca euklidesową odległość pomiędzy punktami 2D.
    """
    pt_1 = np.array((pt_1))
    pt_2 = np.array((pt_2))
    return np.linalg.norm(pt_1 - pt_2, axis = 1)
```

```python
n = 0
for i in range(S2_df.shape[0]):
    print(i)
    S2_df_row = S2_df.iloc[[i]]
    
    weather_stations_df = Get_Weather_Measurements_Data(S2_df['publicationDate'].values[0])
    
    weather_stations_df['im_lat'] = S2_df_row.centroid.values[0][0]
    weather_stations_df['im_lon'] = S2_df_row.centroid.values[0][1]
    
    ### Odleglosc zobraozowania od stacji 
    weather_stations_df['distance'] = distance(weather_stations_df[['lat', 'lon']],
                                                weather_stations_df[['im_lat', 'im_lon']])
    
    nearest_weather_stations = weather_stations_df.sort_values('distance').iloc[[0]]
    
    result = pd.concat([S2_df_row.reset_index(), nearest_weather_stations.reset_index().drop(columns='index')], axis = 1).set_index("id")
    result = result.drop(columns = ['lat', 'lon', 'centroid'])
    data = result.to_dict(orient = 'records')
    db.S2.insert_one(data[0])

    n += 1
```

### 4. Formułowanie zapytań do uzupełnionej bazy danych MongoDB.

```python
### Selekcja zobrazowan w miejscach w których prawdopodobnie wystąpiły przymrozki. 
query_results = db.S2.find({'Temperature[C]': {"$lt": 0}})

query_results_df = pd.io.json.json_normalize(query_results)
query_results_df
```

#### 4.1. Przekształcenie wyników zapytania do ramki geo-danych

```python
# Eliminacje rekorów nie zawierających geometrii
query_results_df = query_results_df[~query_results_df['geometry'].isna()]

# Stworzenie ramki geo-adanych
query_results_df['geometry'] = query_results_df['geometry'].astype(str).apply(wkt.loads)
query_results_gdf = gpd.GeoDataFrame(query_results_df, geometry = query_results_df['geometry'])

# Zdefiniownie ukł. wsp.
query_results_gdf.crs = {"init":"epsg:4326"}
```

### 4.2. Wizualizacja wyników zapytania

```python
### Reprojekcja układu współrzędnych wykorzystywanego przez mapy bazowe internetowych serwisów mapowych - WGS84 / Pseudo Merkator
query_results_gdf = query_results_gdf.to_crs("EPSG:3857")

plt.rcParams["figure.figsize"] = (15, 15)
ax = query_results_gdf.plot(cmap = "viridis",
                                linewidth = 0.1,
                                edgecolor = 'black',
                                legend = True,
                                alpha = 0.2)

cx.add_basemap(ax, source = cx.providers.Stamen.TonerLite)
plt.title("Zasięgi wyników zapytania Sentinel-2B")
plt.show()
```
